## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.
_____________________________________________________________________________________________
The Primary Keys of a table are those columns which are identified uniquely a row.
The Foreign keys are not the Primary Keys of that particular table but for another table.



## Question 3

Explain the different types of relationships entities can have in a SQL database. 
_______________________________________________________________________________________________________________________________
One-to-one: each row of a table is linked to a single row of another table using a key column 
one-to-many: each row of a table is linked to one or more rows of another table using a key column
many-to-many: one or more rows of a table is linked to one or more rows of another table using a key column



## Question 4

Explain the various types of JOINs possible with SQL. 
_______________________________________________________________________________________________________________________________
FULL OUTER JOIN: select all records from Table A and Table B, regardless whether the join condition is met or not
INNER JOIN: select all records from Table A and Table B where the join condition is met
LEFT JOIN: select all records from Table A, along with records from Table B for which the join condition is met (if any)
RIGHT JOIN: select all records from Table B, along with records from Table A for which the join condition is met (if any)




## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.
_____________________________________________________________________________________________
GROUP BY clause is used to output a row accross a specified column value. It is used in conjunction with Aggregate functions to perform a quick calculation on values being returned fronm a SELECT query, in other words Aggregate functions are used to summarize values. 




## Question 6
